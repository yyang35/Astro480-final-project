```python
# importing the required module
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
# importing modules
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization.mpl_normalize import ImageNormalize
from astropy.visualization import SqrtStretch
from matplotlib.patches import Arrow, Circle

def plot_HR(x,y,distance,title,label):
    fig, ax = plt.subplots(figsize=(10, 10))
    
    
    scatter = ax.scatter(x1, y1,s=6, color = "#d4d4d4" )
    scatter1 = ax.scatter(x, y,s=8, c=distance, cmap='viridis')
    scatter = ax.scatter(iso1["Bmag"] - iso1["Vmag"], iso1["Vmag"]+ 15.1, s=30, marker = "_", color = '#FF6759')
    
    plt.colorbar(scatter1,label = 'Distance to center of M3(arcmin)')
    
    
    #ax = scatter.axes
    size = len(x)
    ax.text(-0.25, 13.8, f"Database Size = {size}",fontsize=10)
    #ax.text(1, 12, f"Database Size = {size}")
    ax.yaxis.grid(color='#d4d4d4', linestyle='dashed')
    ax.xaxis.grid(color='#d4d4d4', linestyle='dashed')
    ax.set_facecolor('#fefefe')
    plt.xlabel('B-V',fontsize=10)
    plt.ylabel('V',fontsize=10)
    plt.gca().invert_yaxis()
    
    #ax.scatter([1.247],[13.038], s = 80, color='red', edgecolor='black')
    #ax.text(1.28, 13.158, f"Reference star",color='red', backgroundcolor="white")
    ax.legend(['All Stars Cross Database', label, 'isochrone'], markerscale = 2, fontsize=12)
    plt.title(title,fontsize=12)
    plt.xlim(-0.3,1.6)
    plt.ylim(24,12)
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
    plt.savefig(f"figures/{label}.png",bbox_inches='tight', pad_inches=0)
    #plt.show()
```

```python
def generate(table,title,label):
    result = pd.read_csv(table)
    x = result['Bmag'] - result['Vmag']
    y = result['Vmag']
    distance = result['distance']
    plot_HR(x,y,distance,title,label)
```
