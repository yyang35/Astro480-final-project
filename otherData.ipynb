```python
background = pd.read_csv("UBVRI.csv")
x1 = background['Bmag'] - background['Vmag']
y1 = background['Vmag']
background = background 
```

```python
# importing the required module
import matplotlib.pyplot as plt
def plot_HR(x,y,distance):
    fig, ax = plt.subplots(1,figsize=(10, 8))
    scatter = plt.scatter(x1, y1,s=2, color = "#d4d4d4" )
    scatter = plt.scatter(x, y,s=8, c=distance, cmap='viridis')
    
    ax = scatter.axes

    ax.yaxis.grid(color='#d4d4d4', linestyle='dashed')
    ax.xaxis.grid(color='#d4d4d4', linestyle='dashed')
    ax.set_facecolor('#fefefe')
    plt.xlabel('B-V')
    plt.ylabel('V')
    plt.gca().invert_yaxis()
    plt.colorbar(label = 'Distance to center of M3(arcmin)')
    plt.xlim(-0.3,1.6)
    plt.ylim(24,12)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = merge['Vmag']
x = merge['Gmag']

z = np.polyfit(x=x, y=y, deg=1)
p = np.poly1d(z)
line = p(x)
fig, ax = plt.subplots(1,figsize=(10, 4))
  
#plt.ylim(8,15)
# plotting the points 
plt.plot(x, y, marker='o', fillstyle='none', markersize=4,linestyle = 'none')
plt.plot(x, line, '-', )
plt.legend(["Reference Stars",f'Fitline:{p}'])

#plt.xlabel('G Magnitude')
plt.ylabel('V Magnitude')
#plt.title('Match Up G,V Magnitude')
ax.text(14.2, 16.1, 'Match Up G,V Magnitude', fontsize=12)
ax.text(14.3, 13.4, 'G Magnitude', fontsize=10)
plt.savefig("matchGV.png",bbox_inches='tight', pad_inches=0)

```
